```python
import numpy as np

def get_matrix(name):
    """Helper function to input a matrix from user"""
    rows = int(input(f"Enter number of rows for {name}: "))
    cols = int(input(f"Enter number of columns for {name}: "))
    print(f"Enter elements row by row for {name}:")
    matrix = []
    for i in range(rows):
        row = list(map(float, input().split()))
        if len(row) != cols:
            raise ValueError("Number of elements in row does not match column count")
        matrix.append(row)
    return np.array(matrix)

def matrix_operations():
    print("=== Matrix Operations Tool ===")
    A = get_matrix("Matrix A")
    B = get_matrix("Matrix B")

    print("\nMatrix A:\n", A)
    print("Matrix B:\n", B)

    # Addition
    if A.shape == B.shape:
        print("\nA + B:\n", A + B)
    else:
        print("\nAddition not possible (different dimensions).")

    # Subtraction
    if A.shape == B.shape:
        print("\nA - B:\n", A - B)
    else:
        print("\nSubtraction not possible (different dimensions).")

    # Multiplication
    try:
        print("\nA dot B (matrix multiplication):\n", np.dot(A, B))
    except ValueError:
        print("\nMatrix multiplication not possible (dimension mismatch).")

    # Transpose
    print("\nTranspose of A:\n", A.T)
    print("Transpose of B:\n", B.T)

    # Determinant
    if A.shape[0] == A.shape[1]:
        print("\nDeterminant of A:", np.linalg.det(A))
    else:
        print("\nDeterminant of A not possible (not square).")

    if B.shape[0] == B.shape[1]:
        print("Determinant of B:", np.linalg.det(B))
    else:
        print("Determinant of B not possible (not square).")

# Run tool
matrix_operations()
```
